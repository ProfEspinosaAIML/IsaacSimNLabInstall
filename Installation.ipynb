---

# 2️⃣ Disable Windows GPU virtualization features (if enabled)

Your system report shows **virtualization-based security** and a detected hypervisor. These can sometimes interfere with CUDA and RTX.

Check if **Hyper-V** is enabled.

Open **PowerShell as Administrator** and run:

```powershell
systeminfo | findstr Hyper-V
```

If Hyper-V features are active, disable them temporarily:

```powershell
bcdedit /set hypervisorlaunchtype off
```

Then reboot.

This avoids GPU conflicts that sometimes prevent Isaac Sim from launching.

---

# 3️⃣ Enable long Windows file paths

Isaac Sim downloads thousands of assets and Python packages. Windows’ default **260-character path limit** can break installation.

Enable long paths:

Open **PowerShell (Administrator)** and run:

```powershell
New-ItemProperty `
-Path "HKLM:\SYSTEM\CurrentControlSet\Control\FileSystem" `
-Name "LongPathsEnabled" `
-Value 1 `
-PropertyType DWORD `
-Force
```

Then reboot.

This prevents installation errors when installing Omniverse packages and extensions.

### 3. Simulate Physics

Isaac Sim takes over:

The RL algorithm (PPO — Proximal Policy Optimization):

* Looks at what happened
* Adjusts the neural network
* Improves future decisions

Then the cycle repeats.
